Housing Calculator -

This calculator offers two different calculations. The first calculation is an estimate for how many months it would take the user to save up for the down payment on the house they are looking into getting.
The second calculation is a percentage of how much of the user's monthly income would have to be saved in order for the user to reach their goal in x amount of months. 

This function definition is the first calculation that the user can perform. This calculation will give the user the number of months it would take for them to save up for the down payment on the house they want. With this function definition the user is able to put in whole numbers and decimals. Since it is more of a simple calculation compared to getting an exact percentage, the whole numbers will not have an impact on the result that the user receives. 

```python
def months_to_down_payment(
    total_cost,
    portion_down_payment,
    annual_salary,
    portion_saved,
    r
):
    """
    Calculate the number of months required to save enough money for a down payment on a house. 

    Arguments:
        total_cost (float): Total cost of the home the user is looking to buy.
        portion_down_payment (float): A fraction of the cost that is required for the down payment. 
        annual_salary (float): This is the users annual salary.
        portion_saved (float): This is a percentage of the money from the monthly salary that the user wants to save. 
        r (float): This is the annual rate of return on the user's savings/investments. 

    Returns:
        int: The number of months required to reach the down payment.
    """

    # -- Guardrail --
    if r < 0  or r > 15:
        raise ValueError(
            "Annual rate of return must be between 0 and 15 percent."
        )

    # -- Core Logic -- 
    monthly_salary = annual_salary / 12
    down_payment = total_cost * portion_down_payment

    current_savings = 0.0
    months = 0 

    while current_savings < down_payment:
        current_savings += current_savings * r / 12
        current_savings += monthly_salary * portion_saved 
        months += 1
    
    return months
```

This function definition will give the user the percentage of their monthly income that they would need to save in order to reach their goal within a given number of months. Since this function definition is more specific and gives the user an exact percentage that they will need to save from their monthly income, the user will need to input decimals instead of the whole number. If whole numbers are input the user will receive a ValueError message. 

```python
def required_savings_rate(
    total_cost,
    portion_down_payment,
    annual_salary,
    monthly_goal,
    r
):
    """
    Calculate the required monthly savings rate (as a decimal) needed to reach the down payment goal within the given time frame. 

    Arguments:
        total_cost (float): The total cost of the home the user is interested in purchasing.
        portion_down_payment (float): The fraction of the total cost needed for the down payment.
        annual_salary (float): The user's annual salary. 
        monthly_goal (int): Number of months the user wants to have the down payment by.
        r (float): The annual rate of return on the user's savings/investments. 

    Returns:
        float: Required savings rate as a decimal (ex. 0.125 = 12.5%).
    """

    # -- Guardrail -- 
    if monthly_goal <= 0:
        raise ValueError("Monthly goal must be greater than zero.")
    if portion_down_payment <= 0 or portion_down_payment > 1:
        raise ValueError("Down payment portion must be between 0 and 1 (input as a decimal for accurate calculation).")
    if annual_salary <= 0:
        raise ValueError("Annual salary must be greater than zero.")
    if r < 0:
        raise ValueError("Rate of return cannot be negative.")

    down_payment = total_cost * portion_down_payment
    monthly_salary = annual_salary / 12

    low = 0.0
    high = 1.0
    epsilon = 0.0001
    guess = (high + low) / 2.0

    while True:
        current_savings = 0.0
        for month in range(monthly_goal):
            current_savings += current_savings * r / 12
            current_savings += monthly_salary * guess
        if abs(current_savings - down_payment) <= epsilon:
            break
        if current_savings < down_payment:
            low = guess
        else:
            high = guess
        guess = (high + low) / 2.0

    return guess
```

This is the portion of the code where the user input is being collected and the calculation that the user wants to perform will be chosen. 

```python
choice = input(
    "Choose a calculation:\n"
    "1. Months to save for the down payment\n"
    "2. Savings percentage needed to achieve goal\n"
    "Enter option 1 or option 2."
).strip()

if choice == "1":
    total_cost = float(input("Enter the total cost of the home: "))
    portion_down_payment = float(input("Enter the down payment rate (ex. 25%): "))
    annual_salary = float(input("Enter your annual salary: "))
    portion_saved = float(input("Enter the percentage of your monthly income that is being saved (ex. 20%, 30%, 40%): "))
    r = float(input("Enter the annual rate of return on your savings/investments (ex. 3%, 4%, 5%): "))
elif choice == "2":
    total_cost = float(input("Enter the total cost of the home: "))
    portion_down_payment = float(input("Enter the down payment rate (ex. 25%): "))
    annual_salary = float(input("Enter your annual salary: "))
    monthly_goal = int(input("Enter the number of months you want to have the down payment by: "))
    r = float(input("Enter the annual rate of return on your savings/investments (ex. 3%, 4%, 5%): "))
```

This is the execution of the calculations. Depending on which option the user chose (option 1 or option 2) will determine which answer they receive. If they chose option 1, then the result that the user will receive will be the number of months it will take them to get the full down payment. If they chose the second option, then that will result in the user receiving a percentage in which they would need to save from their monthly income in order to reach their goal in a given number of months. If any other input is collected then the user will be prompted to restart and selection option 1 or option 2.

```python
if choice == "1":
    months = months_to_down_payment(
        total_cost,
        portion_down_payment,
        annual_salary,
        portion_saved,
        r
    )
    print("Number of months needed to reach down payment: ", months)

elif choice == "2":
    savings_percentage = required_savings_rate(
        total_cost,
        portion_down_payment,
        annual_salary,
        monthly_goal,
        r
    )
    if savings_percentage >= 1:
        print("This down payment goal is not achievable with the given income and time frame. Consider increasing the timeline or adjusting the target amount.")
    else:
        print(f"Required savings rate: {savings_percentage:.2f} ({savings_percentage * 100:.2f}%)")
else: 
    print("Invalid selection. Please restart and choose option 1 or option 2.")
```
